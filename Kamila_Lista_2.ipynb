image.png

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
valores_recebidos = np.array([160, 184, 241, 149, 180, 161, 132, 202, 160, 139, 149, 177])
valores_processados = np.array([160, 184, 237, 148, 181, 150, 123, 156,126, 104, 124, 140])
d = {"meses": meses, 'recebidos': valores_recebidos, 'processados': valores_processados}
df = pd.DataFrame(data=d)
plt.figure(figsize=(15,5))
sns.lineplot(data=df, x="meses",y = "recebidos", label = "Recebidos")
sns.lineplot(data=df, x="meses",y = "processados", label = "Processados")

```
